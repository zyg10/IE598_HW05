```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.15, random_state=42)
X = df[df.columns[:-1]].values
y = df['Adj_Close'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15, random_state=42)
y_train=y_train.reshape(-1,1)
y_test=y_test.reshape(-1,1)
mm = MinMaxScaler()
sc = StandardScaler()
sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
y_train = mm.fit_transform(y_train)
y_test = mm.transform(y_test)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.svm import SVR
```

## Linear and SVM Without PCA

```python
slr = LinearRegression()
slr.fit(X_train,y_train)
y_pred = slr.predict(X_test)
```

```python
R2_train = slr.score(X_train, y_train)
RMSE_train = np.sqrt(mean_squared_error(y_train,slr.predict(X_train)))
R2_test = slr.score(X_test, y_test)
RMSE_test = np.sqrt(mean_squared_error(y_test, y_pred))
print('The coefficient is : ', slr.coef_)
print('The intercept is:  ', slr.intercept_)
print('R2 training is: ', R2_train)
print('RMSE training : ', RMSE_train)
print('R2 testing is: ', R2_test)
print('RMSE testing is: ', RMSE_test)
```

```python
svr = SVR(kernel='rbf')
svr.fit(X_train,y_train)
y_pred = svr.predict(X_test)

R2_train = svr.score(X_train, y_train)
RMSE_train = np.sqrt(mean_squared_error(y_train,svr.predict(X_train)))
R2_test = svr.score(X_test, y_test)
RMSE_test = np.sqrt(mean_squared_error(y_test, y_pred))

print('The intercept is:  ', svr.intercept_)
print('R2 training is:: ', R2_train)
print(' RMSE training: ', RMSE_train)
print('R2 testing is: ', R2_test)
print('RMSE testing is: ', RMSE_test)





```

## Linear and SVM with PCA

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=3)
X_pca = df[df.columns[:-1]].values
y_pca = df['Adj_Close'].values
X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(X_pca, y_pca, test_size = 0.15, random_state=42)
sc = StandardScaler()
mm = MinMaxScaler()
y_train_pca=y_train_pca.reshape(-1,1)
y_test_pca=y_test_pca.reshape(-1,1)

sc = StandardScaler().fit(X_train_pca)
X_train_std = sc.transform(X_train_pca)
X_test_std = sc.transform(X_test_pca)
y_train = mm.fit_transform(y_train_pca)
y_test = mm.transform(y_test_pca)

X_train_pca=pca.fit_transform(X_train_std)
X_test_pca=pca.fit_transform(X_test_std)
```

```python
slr = LinearRegression()
slr.fit(X_train_pca,y_train_pca)
y_pred_pca = slr.predict(X_test_pca)

R2_train_pca = slr.score(X_train_pca, y_train_pca)
RMSE_train = np.sqrt(mean_squared_error(y_train_pca,slr.predict(X_train_pca)))
R2_test_pca = slr.score(X_test_pca, y_test_pca)
RMSE_test = np.sqrt(mean_squared_error(y_test_pca, y_pred_pca))
print('The coefficient is : ', slr.coef_)
print('The intercept is:  ', slr.intercept_)
print('R2 training is: ', R2_train)
print('RMSE training: ', RMSE_train)
print('R2 testing is: ', R2_test)
print('RMSE testing is: ', RMSE_test)
```

```python
from sklearn.svm import SVR
svr = SVR(kernel='rbf')
svr.fit(X_train_pca,y_train_pca)
y_pred_pca = svr.predict(X_test_pca)

R2_train = svr.score(X_train_pca, y_train_pca)
RMSE_train = np.sqrt(mean_squared_error(y_train_pca,svr.predict(X_train_pca)))
R2_test = svr.score(X_test_pca, y_test_pca)
RMSE_test = np.sqrt(mean_squared_error(y_test_pca, y_pred_pca))

print('The intercept is:  ', svr.intercept_)
print('R2 training is: ', R2_train)
print('RMSE training: ', RMSE_train)
print('R2 testing is: ', R2_test)
print('RMSE testing is: ', RMSE_test)
```

```python

print("Explained ratio are:",pca.explained_variance_ratio_)
```
